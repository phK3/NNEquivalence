```python
from clustering import RecursiveClustering, ClusterTree
from performance import Encoder

import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
import pickle
from timeit import default_timer as timer
import sys
```

# Load Training Data #

```python
df_train = pd.read_csv("mnist_train.csv",header=None)
df_train.head()
```

```python
df_test = pd.read_csv("mnist_test.csv", header=None)
```

```python
df_train_imgs = df_train.drop(0, axis=1)
df_train_imgs = df_train_imgs / 255.
df_train_labels = df_train[0]
```

```python
df_test_imgs = df_test.drop(0, axis=1)
df_test_imgs = df_test_imgs / 255.
df_test_labels = df_test[0]
```

# Hierarchical Clustering #

```python
rc = RecursiveClustering()
```

```python

ct = rc.recursive_cluster(df_train_imgs, df_train_labels, purity=1, verbose=True)
```

```python
rc.calculate_cluster_distances()
clusters = rc.get_leaves()
```

```python
topClusters = sorted(clusters, key=lambda x: -x.density())
```

```python
cntr = 0
clusters10 = []
for cl in topClusters:
    if cl.elements > 1:
        print("{}: density = {}, numElements = {}, avg_dist = {}, dist2other = {}".format(cntr, cl.density(), 
                                                                                          cl.elements, cl.avg_dist, 
                                                                                          cl.distance))
        clusters10.append(cl)
        cntr += 1
        
    if cntr >= 10:
        break   
```

```python
pickle.dump(clusters10, open( "clusters10.pickle", "wb" ) )
```

# Verification of Properties #

```python
reference = "mnist_full_lin.h5"
test = "mnist_full_90p_retrain.h5"
mode = "one_hot_partial_top_1"
```

```python
input_los = [0 for i in range(784)]
input_his = [1 for i in range(784)]
```

```python
enc = Encoder()
enc.encode_equiv(reference, test, input_los, input_his, mode)
```

```python
cl = clusters10[0]

center = cl.center
steps = [3, 0.05, 0.1, 0.2]
```

```python
time_limit = 10 # per verification procedure
models = []
stdout = sys.stdout
for s in steps:
    teststart = timer()
    
    radius = s * cl.distance
    
    name = "mnist_full_step_{}".format(s)
    logfile = name + '.txt'
    sys.stdout = open(logfile, 'w')
    
    enc = Encoder()
    enc.encode_equiv(reference, test, input_los, input_his, mode)
    enc.add_input_radius(center, radius, metric='manhattan')
    
    enc.optimize_constraints()
    
    model = enc.create_gurobi_model()
    models.append(model)
    
    model.setParam('BestObjStop', 20.)
    model.setParam('BestBdStop', 0.01)
    model.setParam('TimeLimit', time_limit)
    
    model.optimize()
    
    now = timer()
    
    sys.stdout = stdout
    print('### {name} finished. Total time elapsed: {t}'.format(name=name, t=now - teststart))
    print('    radius = {r}'.format(r=radius))
    print('    (val, bound) = ({v}, {bd})'.format(v=model.ObjVal, bd=model.ObjBound))
    
    
```
